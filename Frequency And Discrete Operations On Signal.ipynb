```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 1, 100)
signal = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(len(t))
dft = np.fft.fft(signal)
freqs = np.fft.fftfreq(len(signal), t[1] - t[0])
filter_mask = (np.abs(freqs) < 5) 
dft_filtered = dft * filter_mask
signal_filtered = np.fft.ifft(dft_filtered)
plt.figure(figsize=(10, 4))
plt.plot(freqs, np.abs(dft), label='Original DFT')
plt.plot(freqs, np.abs(dft_filtered), label='Filtered DFT')
plt.title('Frequency Domain Representation')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.legend()
plt.show()
plt.figure(figsize=(10, 4))
plt.plot(t, signal, label='Original Signal')
plt.plot(t, signal_filtered.real, label='Filtered Signal', linewidth=2)
plt.title('Time Domain Signals')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 10 * t) + 0.5 * np.random.randn(len(t))
X = np.fft.fft(x)
freqs = np.fft.fftfreq(len(x), t[1] - t[0])
lowpass = np.abs(freqs) < 50
X_filtered = X * lowpass
x_filtered = np.fft.ifft(X_filtered)
fig, (ax0, ax1) = plt.subplots(nrows=2, sharex=True, figsize=(8, 6))
ax0.plot(t, x)
ax0.set(title='Original Signal')
ax1.plot(t, x_filtered.real)
ax1.set(title='Filtered Signal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
t = np.linspace(0, 1, 1000)
x1 = np.sin(2 * np.pi * 10 * t) + 0.5 * np.sin(2 * np.pi * 20 * t)
x2 = np.sin(2 * np.pi * 10 * t + np.pi/4) + 0.5 * np.sin(2 * np.pi * 20 * t + np.pi/4)
X1 = np.fft.fft(x1)
X2 = np.fft.fft(x2)
Xcorr = np.fft.ifft(X1*np.conj(X2))
plt.figure(figsize=(10, 4))
plt.plot(np.real(Xcorr))
plt.xlabel('Sample')
plt.ylabel('Correlation')
plt.title('Cross-Correlation Between Two Signals')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve
filter_size = 10
filter_kernel = np.ones(filter_size) / filter_size
signal_length = 100
signal = np.random.rand(signal_length)
filtered_signal = convolve(signal, filter_kernel, mode='same')
plt.figure(figsize=(10, 4))
plt.plot(signal, label='Original Signal', alpha=0.5)
plt.plot(filtered_signal, label='Filtered Signal', linewidth=2)
plt.title('Signal Before and After Filtering')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve
filter_size = 10
filter_kernel = np.ones(filter_size) / filter_size
signal_length = 100
signal = np.random.rand(signal_length)
freq_response = np.abs(np.fft.fft(filter_kernel, n=1024))
freqs = np.fft.fftfreq(len(freq_response))
plt.figure(figsize=(10, 4))
plt.plot(freqs[:len(freqs)//2], freq_response[:len(freq_response)//2])
plt.title('Frequency Response of Moving Average Filter')
plt.xlabel('Normalized Frequency')
plt.ylabel('Magnitude')
plt.grid()
plt.show()
plt.figure(figsize=(10, 4))
plt.plot(signal, label='Original Signal', alpha=0.3)

for size in [5, 10, 20]:
    kernel = np.ones(size) / size
    filtered = convolve(signal, kernel, mode='same')
    plt.plot(filtered, label=f'Filter size {size}')

plt.title('Effect of Different Filter Sizes')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()
```

```python
#LAB TASK CONVULATIONS
import numpy as np
from scipy.signal import convolve

def convolution_sum(x, h):
    return convolve(x, h, mode='full')
x = np.array([1, 2, 3])
h1 = np.array([1, 1])
h2 = np.array([2, -1])

first_conv = convolution_sum(x, h1)
result1 = convolution_sum(first_conv, h2)

second_conv = convolution_sum(h1, h2)
result2 = convolution_sum(x, second_conv)
print("(x * h1) * h2:", result1)
print("x * (h1 * h2):", result2)
print("Equal:", np.allclose(result1, result2))
```

```python
from scipy.signal import lfilter
a = [1, -4, 4] 
b = [1, -1]      
impulse = np.zeros(50)
impulse[0] = 1
n = np.arange(-20, 26)
h_difference = lfilter(b, a, impulse)
plt.figure(figsize=(12, 4))
plt.stem(n, h_difference[20-20:20+26])
plt.title('Impulse Response (Difference Equation)')
plt.xlabel('n')
plt.ylabel('h[n]')
plt.show()
# Using filter function
h_filter = lfilter(b, a, impulse)

plt.figure(figsize=(12, 4))
plt.stem(n, h_filter[20-20:20+26])
plt.title('Impulse Response (Filter Function)')
plt.xlabel('n')
plt.ylabel('h[n]')
plt.show()

n = np.arange(-20, 26)
x = (5 + 3*np.cos(0.2*np.pi*n) + 4*np.sin(0.6*np.pi*n)) * (n >= 0)
y = lfilter(b, a, x)
plt.figure(figsize=(12, 4))
plt.stem(n, y)
plt.title('System Output for Given Input')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.show()

n = np.arange(200)
x = np.cos(0.2*np.pi*n) + 0.5*np.cos(0.6*np.pi*n)
y = x + 0.1*np.roll(x, 20)  # 0.1x[n-20]
correlation = np.correlate(y, y, mode='full')
lags = np.arange(-len(y)+1, len(y))
plt.figure(figsize=(12, 4))
plt.plot(lags, correlation)
plt.title('Autocorrelation of y[n]')
plt.xlabel('Lag')
plt.ylabel('Correlation')
plt.grid()
peak_positions = np.where(correlation > 0.9*correlation.max())[0]
echo_delay = abs(peak_positions[1] - peak_positions[0])
print(f"Detected echo delay: {echo_delay} samples")
```

```python
import numpy as np
import matplotlib.pyplot as plt
Fs = 1600 
T = 1 / Fs
N = 32
n = np.arange(N)
x = 3 * np.cos(np.pi * n / 8) + 4 * np.sin(5 * np.pi * n / 16) - np.cos(np.pi * n / 2)
Xk = np.fft.fft(x) / N 
frequencies = np.fft.fftfreq(N, d=1/Fs)
Xk_shifted = np.fft.fftshift(Xk)
frequencies_shifted = np.fft.fftshift(frequencies)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.stem(frequencies_shifted, np.abs(Xk_shifted), basefmt=" ")
plt.title('DTFS Magnitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('|X[k]|')
plt.subplot(1, 2, 2)
plt.stem(frequencies_shifted, np.angle(Xk_shifted), basefmt=" ")
plt.title('DTFS Phase Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (radians)')
plt.tight_layout()
plt.show()
```
