```python
import numpy as np
import matplotlib.pyplot as plt
n = range(-2, 6, 1)
y = []
for i in range(len(n)):
    temp = (1 if n[i]>=0 else 0)
    y.append(temp)

print(n)
print(y)

#plotting the graph
plt.stem(n, y)
plt.axis([-2.1, 5.1, -0.1, 1.3])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = range(-2, 6, 1)
y = []
for i in range(len(n)):
    temp = (n[i] if n[i]>=0 else 0)
    y.append(temp)

print(list(n))
print(y)

#Plotting the graph
plt.stem(n, y)
plt.axis([-2.1, 5.1, -0.1, 5.5])
plt.xlabel('  n--->  ')
plt.ylabel('Amplitude ----> ')
plt.title('Unit ramp Signal')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = 0.9
n = range(0, 26, 1)
y = []

for i in range(len(n)):
    temp = a**n[i]
    y.append(temp)

#Plotting the graph
plt.stem(y)
plt.axis([-0.3, 25.1, -0.1, 1.1])
plt.xlabel('  n--->  ')
plt.ylabel('Amplitude ----> ')
plt.title('Exponential signal for 0 < a < 1')
plt.grid(True)
plt.show()
```
